```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# import the raw data
data = pd.read_csv("bert_mills.csv",sep= ",")
data
```

```python
# Features from BERT and GPT do not have word length, word frequency and average distance to k-nearest words.
# These are obtained from the csv files for W2G-Mills. Here is the example usage:
# If you read files of W2G model, do not run this part

data_temp= pd.read_csv("mills_sph_d50.csv",sep= "," )
data["frequencies"]=data_temp["frequencies"]
data["word_length"]=data_temp["word_length"]


from sklearn.neighbors import NearestNeighbors

def average_distance(data,k):
        # Create a NearestNeighbors object
        nbrs = NearestNeighbors(n_neighbors=k)

        # Fit the data to find the nearest neighbors
        nbrs.fit(data.iloc[:,:-3])
        distances, indices = nbrs.kneighbors(data.iloc[:,:-3])
        mean_distances = np.mean(distances, axis=1)
        data["average_distance"]=mean_distances

        return data
data=average_distance(data,5)

y = data["class"]
del data["class"]
data["class"]=y
data
```

```python
# Normalizes the data
def normalizer(data_frame):
     scaler = StandardScaler()
     data2= data_frame.iloc[:,:-1]
     label=data_frame["class"]
     df_scaled = pd.DataFrame(scaler.fit_transform(data2),columns = data2.columns)
     df_scaled["class"]=label
      
     return df_scaled
```

```python
# Correlation of features are obtained. 
data_normalized=normalizer(data)
cor=data_normalized.corr()
```

```python
threshold=0.30 # Determine a threshold according to "cor" in above, 
a=abs(cor["class"])
result=list(a[a>threshold].index)
data_new= data_normalized[result]
data_new.to_csv("w2deneme.csv",sep=",",index=False)   # Change here
data_new
```
